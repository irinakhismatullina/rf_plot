```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
import sklearn.datasets as ds

from rf_plot import plot_random_forest
```

```python
data = ds.load_breast_cancer()

features = data.feature_names
X, y = data.data, data.target
X.shape, y.shape
```

```python
MAX_DEPTH = 6
model = RandomForestClassifier(max_depth=MAX_DEPTH, n_estimators=100)
model.fit(X, y)
```

```python
plot_random_forest(model, features, colors=['red', 'blue'], alpha=0.1)
```

```python
plot_random_forest(model, features, colors=None, alpha=0.1)
```

```python
plot_random_forest(model, features, colors=None, alpha=0.05)
```
